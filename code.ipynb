```python
def run_quiz():
    questions = [
        {
            "question": "What is the capital of France?",
            "options": ["A) London", "B) Berlin", "C) Paris", "D) Madrid"],
            "answer": "C"
        },
        {
            "question": "Which planet is known as the Red Planet?",
            "options": ["A) Earth", "B) Mars", "C) Jupiter", "D) Venus"],
            "answer": "B"
        },
        {
            "question": "Who is known as the poet of the nation?",
            "options": ["A) Allama Iqbal", "B) Sir Syed Ahmed Khan", "C) Liaquat Ali Khan", "D) Chaudhry Rehmat Ali"],
            "answer": "A"
        },
        {
            "question": "What is the largest mammal in the world?",
            "options": ["A) Elephant", "B) Blue Whale", "C) Giraffe", "D) Lion"],
            "answer": "B"
        },
        {
            "question": "What is the capital of Pakistan?",
            "options": ["A) Karachi", "B) Lahore", "C) Islamabad", "D) Sukkur"],
            "answer": "C"
        },
        {
            "question": "Which is the largest city of Pakistan?",
            "options": ["A) Islamabad", "B) Lahore", "C) Karachi", "D) Quetta"],
            "answer": "C"
        },
        {
            "question": "What is the currency of Pakistan?",
            "options": ["A) Rupees", "B) Dollar", "C) Euro", "D) Pound"],
            "answer": "A"
        },
        {
            "question": "Which is the largest province of Pakistan?",
            "options": ["A) Sindh", "B) Balochistan", "C) Punjab", "D) KPK"],
            "answer": "B"
        },

    ]

    score = 0

    print("Welcome to the Quiz Game!\n")

    # Loop through each question
    for i, q in enumerate(questions, start=1):
        print(f"Question {i}: {q['question']}")
        for option in q['options']:
            print(option)
        answer = input("Your answer (A/B/C/D): ").strip().upper()
        if answer == q['answer']:
            print("‚úÖ Correct!\n")
            score += 1
        else:
            print(f"‚ùå Wrong! The correct answer was {q['answer']}\n")

    print(f"Your final score is {score}/{len(questions)}")
    if score == len(questions):
        print("üéâ Excellent! You got them all right!")
    elif score >= len(questions) // 2:
        print("üëç Good job!")
    else:
        print("üí° Keep practicing!")

if __name__ == "__main__":
    run_quiz()
```
